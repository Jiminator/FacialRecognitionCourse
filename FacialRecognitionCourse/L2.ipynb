## Face Detection Function

As the name suggests, face detection refers to searching for a given image (input) with a certain strategy to determine whether it contains a face, and if it contains a face, returns the position information of all faces in the image(Output).



## Description of the face frame
The position of a face is usually represented by a rectangular box. There are many ways to represent a rectangle, such as:

![rects](imgs/FaceRect2.png)

- Upper left coordinate (x, y) and length and width (w, h)
- Center coordinates (cx, cy) and length and width (w, h)
- Upper left coordinate (x1, y1) and lower right coordinate (x2, y2)

In this lesson, we choose the third way to express the information of the rectangular box, that is, the coordinates of the upper left corner and the coordinates of the lower right corner: (x1, y1, x2, y2).

We encapsulate the commonly used face-related functions into the FaceAPI class, which includes the following interface:

Several structures are mentioned above, and the structure information is as follows:
- Point : The attribute has x, y, representing the coordinates of the point
- Rect : The attribute has p1, p2, which represents the upper left corner and the lower right corner.

In this lesson, we will use the detect interface to understand the function of face detection.

In order to help you visually feel the image and face frame information, we provide two interfaces, show(img) and showWithRects(img,rects).
The first interface can visualize the input img, and the second interface will place the rectangle contained in the rects on the img and then visualize it.



```python
from API.FaceAPI_dlib import *
faceAPI = FaceAPI()

img = cv2.imread('data/Curry.png')
show(img)
rects = faceAPI.detect(img)
print('Face Frame: ', rects)
```

Next, visualize the detected face frame onto the original image.

```python
showWithRects(img,rects)  
```

## Finding the biggest face
Generally speaking, there may be a lot of unrelated faces appearing in the background of the photo, and the subject we want to identify should be the one with the largest face in the picture. Can the students find the current scene through loop statements and conditional statements(The largest face of the area) is displayed?


```python
img = cv2.imread('data/group.png')
rects = faceAPI.detect(img)

area = -1
maxRect = None
for rect in rects:
    area_now = (rect.p2.x-rect.p1.x)*(rect.p2.y-rect.p1.y)
    if area_now > area:
        area = area_now
        maxRect = rect
showWithRects(img,[maxRect])  
```
